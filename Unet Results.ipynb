```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
```

```python
path_image = '/application/pi/Desktop/all imp data/Unet testing/'
image_list = os.listdir(path_image)
```

```python
image_list
```

```python
image_list.remove('.ipynb_checkpoints')
```

```python
for i in range(0,5):
    a = np.load(path_image+image_list[2*i])
    b = np.load(path_image+image_list[2*i+1])
    actual =(b[0])
    plt.figure()
    plt.imshow(actual)
    
    predict =(a[0]>0.4)*1
    plt.figure()
    plt.imshow(predict,cmap="gray")
    print('Accuracy',np.sum((actual==predict)*1/(512*512)*100),'%')
    
    intersection = np.logical_and(actual, predict)
    union = np.logical_or(actual, predict)
    iou_score = np.sum(intersection) / np.sum(union)
    print('IOU = ',iou_score)
    print('*******************')
```
