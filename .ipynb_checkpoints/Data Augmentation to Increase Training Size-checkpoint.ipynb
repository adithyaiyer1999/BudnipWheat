# Importing All Pre-augmented Images

```python
import numpy as np
import scripts.dataAugmentation as aug
import os
import matplotlib.pyplot as plt
```

```python
path_image = '/application/pi/Desktop/all imp data/Unet_training/images/'
path_mask = '/application/pi/Desktop/all imp data/Unet_training/masks/'
```

```python
image_list = os.listdir(path_image)
mask_list = os.listdir(path_image)
```

```python
len(image_list)
image_list.remove('.ipynb_checkpoints')
len(mask_list)
```

## Creating Augmented Images and Saving as .npy files

```python
for i in (image_list):
    image = np.load(path_image+i)
    mask = np.load(path_mask+i)
    img = aug.giveImages(image)
    msk = aug.giveImagesForMask(mask)
    for j in range(1,8):
        np.save('/application/pi/Desktop/all imp data/Unet_training/images/transformed_'+i[0:-4]+str(j), img[j])
        np.save('/application/pi/Desktop/all imp data/Unet_training/masks/transformed_'+i[0:-4]+str(j), msk[j])
        
```
