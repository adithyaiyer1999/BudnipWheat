```python
def BRI(img):
    arr = np.zeros((img.shape[0]+1,img.shape[1],img.shape[2]))
    arr[0:img.shape[0]] = img
    arr[img.shape[0]] = (1/img[2]-1/img[4])/img[5]
    return arr
```
